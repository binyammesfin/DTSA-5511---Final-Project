### **2.2 Data Loading and Merging**

```python
# List of stock codes corresponding to the files
stock_codes = ['F', 'JPM', 'MSFT', 'CVX', 'PG', 'BAC', 'XOM']

# List to store individual DataFrames
dataframes = []

# Load the data for each stock and add a stock code column
for stock_code in stock_codes:
    file_path = f"/mnt/data/{stock_code}.csv"
    df = pd.read_csv(file_path)
    df['Stock_Code'] = stock_code  # Add a column for the stock code
    dataframes.append(df)

# Merge all dataframes into a single dataframe
merged_data = pd.concat(dataframes, ignore_index=True)

print(merged_data.head())  # Display the first few rows of the merged dataframe
print(merged_data.shape)   # Display the shape of the merged dataframe
```

```python
# List of stock codes for Dataset 1 (training/validation) and Dataset 2 (testing)
train_val_codes = ['F', 'JPM', 'MSFT', 'CVX', 'PG']
test_codes = ['BAC', 'XOM']

# Lists to store DataFrames for training/validation and testing
train_val_dfs = []
test_dfs = []

# Load and separate the data based on stock codes
for stock_code in train_val_codes:
    file_path = f"./data/{stock_code}.csv"
    df = pd.read_csv(file_path)
    df['Stock_Code'] = stock_code  # Add a column for the stock code
    train_val_dfs.append(df)

for stock_code in test_codes:
    file_path = f"./data/{stock_code}.csv"
    df = pd.read_csv(file_path)
    df['Stock_Code'] = stock_code  # Add a column for the stock code
    test_dfs.append(df)

# Merge the dataframes into two separate DataFrames
train_val_data = pd.concat(train_val_dfs, ignore_index=True)
test_data = pd.concat(test_dfs, ignore_index=True)

print("Training/Validation Data Shape:", train_val_data.shape)
print("Testing Data Shape:", test_data.shape)
```

### **2.3 Data Cleaning and Preprocessing**

In this section, we will handle missing values, normalize the data, and perform feature engineering to create new variables that can enhance the predictive power of our models.

**Handling Missing Values**

The first step in the data cleaning is to address any missing values in the dataset. Missing data can lead to inaccuracies in model predictions, as most machine learning algorithms cannot handle missing values directly. We will handle missing values by using forward-fill methods to carry the last known value forward or by removing rows with missing values if they are sparse enough. This ensures that the dataset is complete and consistent for modeling.

**Normalization**

Normalization is a crucial step in the preprocessing pipeline, especially for models like neural networks. Stock price data often spans different ranges due to varying price scales of different stocks. To ensure that all features are on a common scale, we apply Min-Max scaling. Normalizing the data helps in faster convergence during model training and prevents models from being biased towards features with larger numerical ranges.

**Feature Engineering**

Feature engineering involves creating new variables that can provide additional insights and improve the model's predictive capability. In the context of stock price prediction, certain derived features, such as moving averages and volatility, can be particularly valuable. Moving averages help smooth out short-term fluctuations and highlight longer-term trends, while volatility measures can capture the degree of variation in stock prices over a given period.

**Training/Validation Data Preprocessing**

```python
# Handling missing values
train_val_data.fillna(method='ffill', inplace=True)
train_val_data.dropna(inplace=True)

# Normalization
scaler = MinMaxScaler(feature_range=(0, 1))
train_val_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] = scaler.fit_transform(train_val_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']])

# Feature Engineering
train_val_data['Moving_Avg_50'] = train_val_data.groupby('Stock_Code')['Close'].transform(lambda x: x.rolling(window=50).mean())
train_val_data['Moving_Avg_200'] = train_val_data.groupby('Stock_Code')['Close'].transform(lambda x: x.rolling(window=200).mean())
train_val_data['Volatility'] = train_val_data.groupby('Stock_Code')['Close'].transform(lambda x: x.rolling(window=50).std())
```

```python
train_val_data.tail()
```

**Testing Data Preprocessing**

```python
# Handling missing values
test_data.fillna(method='ffill', inplace=True)
test_data.dropna(inplace=True)

# Normalization
test_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] = scaler.transform(test_data[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']])

# Feature Engineering
test_data['Moving_Avg_50'] = test_data.groupby('Stock_Code')['Close'].transform(lambda x: x.rolling(window=50).mean())
test_data['Moving_Avg_200'] = test_data.groupby('Stock_Code')['Close'].transform(lambda x: x.rolling(window=200).mean())
test_data['Volatility'] = test_data.groupby('Stock_Code')['Close'].transform(lambda x: x.rolling(window=50).std())
```

### **4.1 Auto Regression (AR)**

**Characteristics of the AR Model**

The Auto Regression (AR) model predicts future values based on a linear relationship with past values. It’s denoted as AR(p), where p represents the number of previous time steps (lags) used in the model. The simplicity of the AR model makes it a strong baseline in time series forecasting, especially when data shows autocorrelation.

**Reason for Using the AR Model**

The AR model is ideal for stock price prediction because it effectively captures the influence of past prices on future prices. It provides a straightforward baseline for comparison with more complex models and is easy to interpret, helping to understand basic patterns in the data.

In this project, the AR model will be implemented to serve as a foundational tool for forecasting and as a benchmark for more advanced modeling techniques.

```python
# AR model for a specific stock, e.g., Ford (F)
f_data = train_val_data[train_val_data['Stock_Code'] == 'F']['Close']

# Split the data into training and test sets
train_size = int(len(f_data) * 0.8)
train, test = f_data[:train_size], f_data[train_size:]

# Fit the AR model
ar_model = AutoReg(train, lags=5).fit()

# Make predictions
ar_predictions = ar_model.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

# Calculate performance metrics
ar_mae = mean_absolute_error(test, ar_predictions)
ar_rmse = np.sqrt(mean_squared_error(test, ar_predictions))
ar_mape = np.mean(np.abs((test - ar_predictions) / test)) * 100

print(f"AR Model - MAE: {ar_mae}, RMSE: {ar_rmse}, MAPE: {ar_mape}")
```

### **4.2 Auto Regressive Moving Average (ARMA)**

**Characteristics of the ARMA Model**

The Auto Regressive Moving Average (ARMA) model combines two key components: Auto Regression (AR) and Moving Average (MA). The AR component models the dependency of the current value on its past values, while the MA component captures the influence of past forecast errors. The model is denoted as ARMA(p, q), where p is the number of AR terms, and q is the number of MA terms.

**Reason for Using the ARMA Model**

The ARMA model is used because it balances the strengths of both AR and MA components, making it effective for capturing more complex time series patterns, such as short-term fluctuations and long-term trends. It’s particularly useful when the data shows both autocorrelation and a pattern in the residuals (errors).

In this project, the ARMA model will be employed to improve upon the AR model by incorporating past forecast errors, providing a more nuanced and accurate prediction of stock prices. It will serve as an intermediate step between the simple AR model and more advanced techniques, offering a deeper understanding of the data's dynamics.

```python
# ARMA model for Ford (F)
arma_model = ARIMA(train, order=(5, 0, 5)).fit()

# Make predictions
arma_predictions = arma_model.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

# Calculate performance metrics
arma_mae = mean_absolute_error(test, arma_predictions)
arma_rmse = np.sqrt(mean_squared_error(test, arma_predictions))
arma_mape = np.mean(np.abs((test - arma_predictions) / test)) * 100

print(f"ARMA Model - MAE: {arma_mae}, RMSE: {arma_rmse}, MAPE: {arma_mape}")
```

<a id="4.3"></a>
### **4.3 Auto Regressive Integrated Moving Average (ARIMA)**

**Characteristics of the ARIMA Model**

The Auto Regressive Integrated Moving Average (ARIMA) model extends the ARMA model by including an integration component, which makes the time series data stationary by differencing it. The model is denoted as ARIMA(p, d, q), where p is the number of AR terms, d is the number of differencing steps needed to make the series stationary, and q is the number of MA terms.

**Reason for Using the ARIMA Model**

The ARIMA model is particularly powerful for time series that exhibit non-stationarity, where the mean and variance change over time. By integrating the differencing step, ARIMA can effectively model and forecast data that have trends or seasonal patterns.

In this project, the ARIMA model will be used to handle stock price data that may not be stationary, providing a more flexible and robust approach to predicting future prices. It builds upon the ARMA model by addressing non-stationarity, making it suitable for more complex time series patterns.

```python
# ARIMA model for Ford (F)
arima_model = ARIMA(train, order=(5, 1, 5)).fit()

# Make predictions
arima_predictions = arima_model.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

# Calculate performance metrics
arima_mae = mean_absolute_error(test, arima_predictions)
arima_rmse = np.sqrt(mean_squared_error(test, arima_predictions))
arima_mape = np.mean(np.abs((test - arima_predictions) / test)) * 100

print(f"ARIMA Model - MAE: {arima_mae}, RMSE: {arima_rmse}, MAPE: {arima_mape}")
```

### **4.4 Vanilla RNN**

**Characteristics of the Vanilla RNN Model**

A Vanilla Recurrent Neural Network (RNN) is a type of neural network specifically designed for sequential data, such as time series. Unlike traditional neural networks, RNNs have connections that loop back, allowing them to retain information from previous time steps. This makes RNNs capable of capturing temporal dependencies in data.

**Reason for Using the Vanilla RNN Model**

The Vanilla RNN is used for its ability to model sequential data by considering the order and context of past events, which is crucial in time series forecasting like stock prices. It can capture patterns that are not just based on a fixed number of past observations, but also on the sequence of those observations.

In this project, the Vanilla RNN will be applied to leverage its strength in handling temporal dependencies, providing a more dynamic and context-aware prediction of stock prices. It serves as a foundation for more advanced recurrent models and offers insights into the temporal structure of the data.

```python
# Reshape data for RNN
X_train = np.array(train[:-10]).reshape(-1, 1, 1)
y_train = np.array(train[10:]).reshape(-1, 1)
X_test = np.array(test[:-10]).reshape(-1, 1, 1)
y_test = np.array(test[10:]).reshape(-1, 1)

# Building the Vanilla RNN model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(50, input_shape=(X_train.shape[1], X_train.shape[2])))
rnn_model.add(Dense(1))
rnn_model.compile(optimizer='adam', loss='mean_squared_error')

# Training the RNN model
rnn_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)

# Make predictions
rnn_predictions = rnn_model.predict(X_test).flatten()

# Calculate performance metrics
rnn_mae = mean_absolute_error(y_test, rnn_predictions)
rnn_rmse = np.sqrt(mean_squared_error(y_test, rnn_predictions))
rnn_mape = np.mean(np.abs((y_test.flatten() - rnn_predictions) / y_test.flatten())) * 100

print(f"RNN Model - MAE: {rnn_mae}, RMSE: {rnn_rmse}, MAPE: {rnn_mape}")
```

<a id="4.5"></a>
### **4.5 Long Short-Term Memory (LSTM)**

**Characteristics of the LSTM Model**

Long Short-Term Memory (LSTM) is a specialized type of Recurrent Neural Network (RNN) designed to better capture long-term dependencies in sequential data. LSTMs address the limitations of standard RNNs by using gating mechanisms (input, forget, and output gates) to control the flow of information, allowing the model to retain or forget information over long sequences of time steps.

**Reason for Using the LSTM Model**

LSTM is particularly useful for time series data where the relationship between past and future values extends over long periods. In stock price prediction, this ability to remember important information across many time steps helps LSTM models capture trends and patterns that are crucial for accurate forecasting.

In this project, LSTM will be used to enhance the model's ability to learn long-term dependencies in stock prices, providing a more powerful and accurate prediction tool compared to standard RNNs. Its advanced memory capabilities make it especially suitable for complex time series data.

```python
# Building the LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
lstm_model.add(LSTM(50))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')

# Training the LSTM model
lstm_model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=0)

# Make predictions
lstm_predictions = lstm_model.predict(X_test).flatten()

# Calculate performance metrics
lstm_mae = mean_absolute_error(y_test, lstm_predictions)
lstm_rmse = np.sqrt(mean_squared_error(y_test, lstm_predictions))
lstm_mape = np.mean(np.abs((y_test.flatten() - lstm_predictions) / y_test.flatten())) * 100

print(f"LSTM Model - MAE: {lstm_mae}, RMSE: {lstm_rmse}, MAPE: {lstm_mape}")
```

### **4.8 Hyperparameter Tuning**

Hyperparameter tuning is a crucial step in optimizing the performance of machine learning models, especially complex ones like Long Short-Term Memory (LSTM) networks. Unlike model parameters, which are learned during the training process, hyperparameters are set before the training begins and control various aspects of the model's learning process, such as the architecture and training dynamics.

For the LSTM model, hyperparameter tuning involves systematically searching for the best combination of settings that will yield the highest predictive accuracy on the stock price prediction task. Key hyperparameters to be tuned include:

- **Number of LSTM Units**: Controls the complexity of the model by determining the number of neurons in each LSTM layer. More units can capture more intricate patterns, but may also lead to overfitting.
- **Number of Layers**: Stacking multiple LSTM layers can enable the model to learn higher-level representations of the data, but can also increase computational complexity.
- **Learning Rate**: Governs the speed at which the model updates its weights during training. A lower learning rate might lead to better convergence but requires more training time, while a higher learning rate might converge faster but risks overshooting optimal values.
- **Batch Size**: Determines the number of samples the model processes before updating its weights. Smaller batch sizes can lead to more stable training but are computationally more expensive.
- **Dropout Rate**: Helps prevent overfitting by randomly dropping units during training. Tuning the dropout rate is critical for balancing model performance and generalization.

The goal of this hyperparameter tuning process is to find the optimal set of these hyperparameters that will allow the LSTM model to achieve the lowest error rates and the most accurate predictions on unseen stock price data. We will use techniques like grid search or random search to explore different combinations of hyperparameters, ultimately selecting the best configuration based on validation performance.

```python
# Function to create the LSTM model
def create_lstm_model(units, dropout_rate, learning_rate):
    model = Sequential()
    model.add(LSTM(units, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dropout(dropout_rate))
    model.add(LSTM(units))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    
    # Compile the model
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')
    
    return model
```

```python
# Reshape data for LSTM
X_train = np.array(train[:-10]).reshape(-1, 1, 1)
y_train = np.array(train[10:]).reshape(-1, 1)
```

```python
# Define the Hyperparameter Grid
param_grid = {
    'units': [50, 100, 150],
    'dropout_rate': [0.2, 0.3, 0.4],
    'learning_rate': [0.001, 0.01],
    'batch_size': [32, 64],
    'epochs': [50, 100]
}
```

```python
# Train model
results = []

# Create all combinations of hyperparameters
keys, values = zip(*param_grid.items())
param_combinations = [dict(zip(keys, v)) for v in itertools.product(*values)]

# Loop through each combination of hyperparameters
for params in param_combinations:
    print(f"Training with params: {params}")
    
    # Create the model with the current set of hyperparameters
    model = create_lstm_model(params['units'], params['dropout_rate'], params['learning_rate'])
    
    # Train the model
    history = model.fit(X_train, y_train, epochs=params['epochs'], batch_size=params['batch_size'], verbose=0)
    
    # Predict on the training data
    predictions = model.predict(X_train)
    
    # Calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_train, predictions))
    
    # Store the results
    results.append({
        'units': params['units'],
        'dropout_rate': params['dropout_rate'],
        'learning_rate': params['learning_rate'],
        'batch_size': params['batch_size'],
        'epochs': params['epochs'],
        'RMSE': rmse
    })

# Convert results to a DataFrame for easy analysis
results_df = pd.DataFrame(results)
print(results_df)

# Save the results for further analysis
results_df.to_csv("manual_lstm_hyperparameter_tuning_results.csv", index=False)
```

```python
# Display the best result
best_result = results_df.loc[results_df['RMSE'].idxmin()]
print("Best Parameters: ", best_result)
```

Based on the results of the hyperparameter tuning, the LSTM model with the following configuration—50 units, a dropout rate of 0.2, a learning rate of 0.001, a batch size of 32, and 100 epochs—achieved an RMSE of 0.002187. This represents an improvement over the initial LSTM model, which had an RMSE of 0.002285, along with an MAE of 0.001589 and a MAPE of 7.237. Given this improvement, the tuned LSTM model demonstrates a better ability to minimize prediction errors and is therefore the preferred model to apply to the test dataset. This model's optimized parameters are likely to generalize better and provide more accurate predictions on unseen data.

### **4.10 Future Work and Improvements**

Based on the results from both the validation and test datasets, the following areas can be targeted for future improvements:

**Regularization Techniques**:

- **Increase Dropout Rate**: To prevent overfitting, experimenting with a higher dropout rate could help the model generalize better to unseen data.
- **L2 Regularization**: Adding L2 regularization to the model's layers might reduce overfitting by penalizing large weights.

**Advanced Hyperparameter Tuning**:

- **Bayesian Optimization**: Using more sophisticated hyperparameter optimization techniques like Bayesian Optimization could potentially find better hyperparameter combinations more efficiently than grid search.
- **Longer Search Range** Extending the range of hyperparameters such as learning rate, number of units, and batch size could help discover more optimal configurations.

**Model Architecture Enhancements**:

- **Deeper LSTM Layers**: Adding more layers to the LSTM network could allow the model to capture more complex patterns in the data.
- **Bidirectional LSTM**: Implementing a bidirectional LSTM might improve performance by allowing the model to learn from both past and future data sequences.

**Ensemble Methods**:

- **Model Averaging**: Combining predictions from multiple LSTM models with different hyperparameters could stabilize and improve prediction accuracy.
- **Hybrid Models**: Integrating LSTM with other models like ARIMA or a convolutional neural network (CNN) could enhance predictive power by capturing different aspects of the time series data.

**Feature Engineering**:

- **Incorporate Additional Features**: Including more features such as moving averages of different periods, market sentiment indicators, or macroeconomic variables might improve model accuracy.
- **Feature Selection**: Using techniques like Principal Component Analysis (PCA) or Recursive Feature Elimination (RFE) to select the most relevant features could reduce noise and enhance model performance.

**Data Augmentation**:

- **Synthetic Data Generation**: Creating additional synthetic data points using techniques like SMOTE (Synthetic Minority Over-sampling Technique) could help the model learn better and handle variations in the data.
- **Cross-Validation on Different Time Periods**: Performing cross-validation across different time periods to ensure the model is not overfitting to a specific market condition.

**Evaluation Metrics**:

- **Incorporate Additional Metrics**: Using other evaluation metrics like R-squared or Huber loss might provide a more nuanced view of model performance and guide further improvements.

By focusing on these areas, future iterations of the model could achieve better generalization, improved accuracy, and greater robustness in predicting stock prices.
